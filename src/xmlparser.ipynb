```python
def SearchRecursive(ordered_dict, key):
    for k, v in ordered_dict.items():
        if isinstance(v, OrderedDict):
            rec = SearchRecursive(v, key)
            if rec is not None:
                return rec
        if k == key:
            return v
    return None

def SearchUnrecursive(ordered_dict, key):
    for k, v in ordered_dict.items():
        if k == key:
            return v
    return None
```

```python
def ReadInfoFromGroup(filename: str):
    with open(filename, 'r', encoding='utf-8') as f:
        my_xml = f.read()
    origin_dict = xmltodict.parse(my_xml)["item"]
    new_dict = {}
    # who info
    try:
        new_dict["who"] = SearchUnrecursive(origin_dict, "who")
        if new_dict["who"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value who is not found in {filename}")

    # rating info
    try:
        new_dict["rating"] = SearchUnrecursive(origin_dict, "rating")
        if new_dict["rating"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value rating is not found in {filename}")

    # topics info
    try:
        new_dict["topics"] = dict(SearchUnrecursive(origin_dict, "topics"))["topics_item"]
        if not isinstance(new_dict["topics"], list):
            new_dict["topics"] = [dict(SearchUnrecursive(origin_dict, "topics"))["topics_item"]]
        if new_dict["topics"] is None:
            raise TypeError
    except:
        new_dict["topics"] = []
        # raise ValueError(f"Value topics is not found in {filename}")

    # category info
    try:
        new_dict["category"] = SearchUnrecursive(origin_dict, "category")
        if new_dict["category"] is None:
            raise TypeError
    except:
        new_dict["category"] = "unknown"
        # raise ValueError(f"Value category is not found in {filename}")

    # categoryname info
    try:
        new_dict["categoryname"] = SearchUnrecursive(origin_dict, "category")["name"]
        if new_dict["categoryname"] is None:
            raise TypeError
    except:
        new_dict["categoryname"] = "unknown"
        # raise ValueError(f"Value categoryname is not found in {filename}")

    # join_mode info
    try:
        new_dict["join_mode"] = SearchUnrecursive(origin_dict, "join_mode")
        if new_dict["join_mode"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value join_mode is not found in {filename}")

    # lat info
    try:
        new_dict["lat"] = SearchUnrecursive(origin_dict, "lat")
        if new_dict["lat"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value lat is not found in {filename}")

    # lon info
    try:
        new_dict["lon"] = SearchUnrecursive(origin_dict, "lon")
        if new_dict["lon"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value lon is not found in {filename}")

    # id info
    try:
        new_dict["id"] = SearchUnrecursive(origin_dict, "id")
        if new_dict["id"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value id is not found in {filename}")

    # created info
    try:
        new_dict["created"] = SearchUnrecursive(origin_dict, "created")
        if new_dict["created"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value created is not found in {filename}")

    # name info
    try:
        new_dict["name"] = SearchUnrecursive(origin_dict, "name")
        if new_dict["name"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value name is not found in {filename}")

    # members info
    try:
        new_dict["members"] = SearchUnrecursive(origin_dict, "members")
        if new_dict["members"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value members is not found in {filename}")

    # country info
    try:
        new_dict["country"] = SearchUnrecursive(origin_dict, "country")
        if new_dict["country"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value country is not found in {filename}")

    # organizer info
    try:
        new_dict["organizer"] = SearchUnrecursive(origin_dict, "organizer")["member_id"]
        if new_dict["organizer"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value organizer is not found in {filename}")

    # urlname info
    try:
        new_dict["urlname"] = SearchUnrecursive(origin_dict, "urlname")
        if new_dict["urlname"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value urlname is not found in {filename}")

    # city info
    try:
        new_dict["city"] = SearchUnrecursive(origin_dict, "city")
        if new_dict["city"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value city is not found in {filename}")

    # link info
    try:
        new_dict["link"] = SearchUnrecursive(origin_dict, "link")
        if new_dict["link"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value link is not found in {filename}")
    return new_dict
```

```python
def ReadInfoFromMember(filename: str):
    with open(filename, 'r', encoding='utf-8') as f:
        my_xml = f.read()
    origin_dict = xmltodict.parse(my_xml)["member"]
    new_dict = {}
    # joined info
    try:
        new_dict["joined"] = SearchUnrecursive(origin_dict, "joined")
        if new_dict["joined"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value joined is not found in {filename}")

    # lat info
    try:
        new_dict["lat"] = SearchUnrecursive(origin_dict, "lat")
        if new_dict["lat"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value lat is not found in {filename}")

    # lon info
    try:
        new_dict["lon"] = SearchUnrecursive(origin_dict, "lon")
        if new_dict["lon"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value lon is not found in {filename}")

    # id info
    try:
        new_dict["id"] = SearchUnrecursive(origin_dict, "id")
        if new_dict["id"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value id is not found in {filename}")

    # topics info
    try:
        new_dict["topics"] = dict(SearchUnrecursive(origin_dict, "topics"))["topics_item"]
        if not isinstance(new_dict["topics"], list):
            new_dict["topics"] = [dict(SearchUnrecursive(origin_dict, "topics"))["topics_item"]]
        if new_dict["topics"] is None:
            raise TypeError
    except:
        new_dict["topics"] = []
        # raise ValueError(f"Value topics is not found in {filename}")

    # hometown info
    try:
        new_dict["hometown"] = SearchUnrecursive(origin_dict, "hometown")
        if new_dict["hometown"] is None:
            raise TypeError
    except:
        new_dict["hometown"] = "unkown"
        # raise ValueError(f"Value hometown is not found in {filename}")

    # name info
    try:
        new_dict["name"] = SearchUnrecursive(origin_dict, "name")
        if new_dict["name"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value name is not found in {filename}")

    # lang info
    try:
        new_dict["lang"] = SearchUnrecursive(origin_dict, "lang")
        if new_dict["lang"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value lang is not found in {filename}")

    # country info
    try:
        new_dict["country"] = SearchUnrecursive(origin_dict, "country")
        if new_dict["country"] is None:
            raise TypeError
    except:
        new_dict["country"] = "unknown"

    # city info
    try:
        new_dict["city"] = SearchUnrecursive(origin_dict, "city")
        if new_dict["city"] is None:
            raise TypeError
    except:
        new_dict["city"] = "unknown"

    # link info
    try:
        new_dict["link"] = SearchUnrecursive(origin_dict, "link")
        if new_dict["link"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value link is not found in {filename}")

    # visited info
    try:
        new_dict["visited"] = SearchUnrecursive(origin_dict, "visited")
        if new_dict["visited"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value visited is not found in {filename}")
    return new_dict
```

```python
def ReadInfoFromEvent(filename: str):
    with open(filename, 'r', encoding='utf-8') as f:
        my_xml = f.read()
    origin_dict = xmltodict.parse(my_xml)["item"]
    new_dict = {}
    # venue info
    try:
        new_dict["venue"] = SearchUnrecursive(origin_dict, "venue")
        if new_dict["venue"] is None:
            raise TypeError
    except:
        new_dict["venue"] = "virtual"
        # raise ValueError(f"Value venue is not found in {filename}")

    # lat info
    try:
        new_dict["lat"] = SearchRecursive(origin_dict, "lat")
        if new_dict["lat"] is None:
            raise TypeError
    except:
        new_dict["venue"] = "virtual"
        # raise ValueError(f"Value lat is not found in {filename}")

    # lon info
    try:
        new_dict["lon"] = SearchRecursive(origin_dict, "lon")
        if new_dict["lon"] is None:
            raise TypeError
    except:
        new_dict["venue"] = "virtual"
        # raise ValueError(f"Value lon is not found in {filename}")

    # id info
    try:
        new_dict["id"] = SearchUnrecursive(origin_dict, "id")
        if new_dict["id"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value id is not found in {filename}")

    # event_hosts info
    try:
        new_dict["event_hosts"] = SearchUnrecursive(origin_dict, "event_hosts")
        if new_dict["event_hosts"] is None:
            raise TypeError
    except:
        new_dict["venue"] = "virtual"
        # raise ValueError(f"Value event_hosts is not found in {filename}")

    # group info
    try:
        new_dict["group"] = SearchUnrecursive(origin_dict, "group")
        if new_dict["group"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value group is not found in {filename}")

    # created info
    try:
        new_dict["created"] = SearchUnrecursive(origin_dict, "created")
        if new_dict["created"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value created is not found in {filename}")

    # visibility info
    try:
        new_dict["visibility"] = SearchUnrecursive(origin_dict, "visibility")
        if new_dict["visibility"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value visibility is not found in {filename}")

    # time info
    try:
        new_dict["time"] = SearchUnrecursive(origin_dict, "time")
        if new_dict["time"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value time is not found in {filename}")

    # event_url info
    try:
        new_dict["event_url"] = SearchUnrecursive(origin_dict, "event_url")
        if new_dict["event_url"] is None:
            raise TypeError
    except:
        raise ValueError(f"Value event_url is not found in {filename}")

    return new_dict
```

```python
def ReadInfoFromRSVP(filename: str):
    with open(filename, 'r', encoding='utf-8') as f:
        my_xml = f.read()
    if xmltodict.parse(my_xml).get("results") is None:
        return None
    if xmltodict.parse(my_xml)["results"].get("items") is None:
        return None
    origin_list = xmltodict.parse(my_xml)["results"]["items"]["item"]
    new_dict = {}

    for origin_dict_item in origin_list:
        useless_count = 0
        # Only fetch in the first person
        try:
            new_dict["venue"] = SearchUnrecursive(origin_dict_item, "venue")
            if new_dict["venue"] is None:
                raise TypeError
        except:
            new_dict["venue"] = "virtual"
            useless_count += 1
            # raise ValueError(f"Value venue is not found in {filename}")
        try:
            new_dict["event"] = SearchUnrecursive(origin_dict_item, "event")
            if new_dict["event"] is None:
                raise TypeError
        except:
            new_dict["event"] = {"id": -1, "name": "virtual"}
            useless_count += 1
            # raise ValueError(f"Value event is not found in {filename}")
        try:
            new_dict["group"] = SearchUnrecursive(origin_dict_item, "group")
            if new_dict["group"] is None:
                raise TypeError
        except:
            new_dict["group"] = {"id": -1, "name": "virtual"}
            useless_count += 1
            # raise ValueError(f"Value group is not found in {filename}")

        if useless_count == 3:
            return None
        break

    memberinfo = []
    for origin_dict_item in origin_list:
        current_member = None
        try:
            current_member = SearchUnrecursive(origin_dict_item, "member")
            if current_member is None:
                raise TypeError
        except:
            raise ValueError(f"Value member is not found in {filename}")
        try:
            current_member["mtime"] = (SearchUnrecursive(origin_dict_item, "mtime"))
            if current_member["mtime"] is None:
                raise TypeError
        except:
            raise ValueError(f"Value mtime is not found in {filename}")

        memberinfo.append(current_member)

    new_dict["memberinfo"] = memberinfo
    return new_dict
```

```python
def TopicsExtractor():
    member_info_file_list = "../data/member.txt"
    print("Checking Members")
    topics_dict = {}
    file_list = []
    with open(member_info_file_list, "r") as f:
        for line in f.readlines():
            line = line.strip()
            file_list.append(line)
    file_index = 1
    file_number = len(file_list)
    for filename in file_list:
        print(f"Dealing with {file_index}/{file_number} ...", end="\r")
        file_index += 1
        info_dict = ReadInfoFromMember(filename)
        if len(info_dict["topics"]) == 0:
            continue
        for item in info_dict["topics"]:
            topics_dict[item["id"]] = item["name"]
    print("Done", end="\r\n")
    group_info_file_list = "../data/group.txt"
    print("Checking Groups")
    file_list = []
    with open(group_info_file_list, "r") as f:
        for line in f.readlines():
            line = line.strip()
            file_list.append(line)
    file_index = 1
    file_number = len(file_list)
    for filename in file_list:
        print(f"Dealing with {file_index}/{file_number} ...", end="\r")
        file_index += 1
        info_dict = ReadInfoFromGroup(filename)
        if len(info_dict["topics"]) == 0:
            continue
        for item in info_dict["topics"]:
            topics_dict[item["id"]] = item["name"]

    return topics_dict
```

```python
def PrintcsvTopics():
    topics_dict = TopicsExtractor()

    with open("../data/topics.csv", 'w') as csvfile:
        csvfile.write("id, name\n")
        for index in range(len(topics_dict)):
            key = list(topics_dict)[index]
            val = topics_dict[key]
            csvfile.write(f"{key}, {val}\n")

PrintcsvTopics()
```

```python
def FormatInfoString(s : str) -> str:
    result = s
    result = result.replace('\'','_s')
    result = result.replace('\"','__s')
    return result
```

```python
def PrintcsvGroups():
    group_info_file_list = "../data/group.txt"
    csvfilename = "../data/group.csv"

    print("Printing out Groups")
    file_list = []
    with open(group_info_file_list, "r") as f:
        for line in f.readlines():
            line = line.strip()
            file_list.append(line)
    file_index = 1
    file_number = len(file_list)

    csv_heading = [
        "id",
        "name",
        "urlname",
        "rating",
        "categoryname",
        "join_mode",
        "lat",
        "lon",
        "country",
        "city",
        "created",
        "members",
        "organizer"
    ]

    with open("../data/group.csv", "w") as csvfile:
        csvfile.write(", ".join(csv_heading) + "\n")
        for filename in file_list:
            print(f"Dealing with {file_index}/{file_number} ...", end="\r")
            file_index += 1
            info_dict = ReadInfoFromGroup(filename)

            current_info = []
            for item in csv_heading:
                current_info.append(FormatInfoString(info_dict[item]))
            csvfile.write("\"" + "\", \"".join(current_info) + "\"" + "\n")

    csv_heading = [
        "groupid",
        "topicid"
    ]

    print("Listing out Group-Topic Relations")
    file_index = 1
    with open("../data/grouptopics.csv", "w") as csvfile:
        csvfile.write(", ".join(csv_heading) + "\n")
        for filename in file_list:
            print(f"Dealing with {file_index}/{file_number} ...", end="\r")
            file_index += 1
            info_dict = ReadInfoFromGroup(filename)
            group_id = info_dict["id"]
            if len(info_dict["topics"]) == 0:
                continue
            for item in info_dict["topics"]:
                topic_id = item["id"]
                csvfile.write(f"{group_id}, {topic_id}\n")

PrintcsvGroups()
```

```python
def PrintcsvEvents():
    group_info_file_list = "../data/event.txt"

    print("Printing out Events")
    file_list = []
    with open(group_info_file_list, "r") as f:
        for line in f.readlines():
            line = line.strip()
            file_list.append(line)
    file_index = 1
    file_number = len(file_list)

    csv_heading = [
        "id",
        "name",
        "event_url",
        "lat",
        "lon",
        "country",
        "city",
        "groupid",
        "created",
        "visibility",
        "time",
    ]

    with open("../data/event.csv", "w") as csvfile:
        csvfile.write(", ".join(csv_heading) + "\n")
        for filename in file_list:
            print(f"Dealing with {file_index}/{file_number} ...", end="\r")
            file_index += 1
            info_dict = ReadInfoFromEvent(filename)

            current_info = [info_dict["id"]]
            if isinstance(info_dict["venue"], str) or info_dict["venue"] is None:
                current_info.append("virtual")
                current_info.append(info_dict["event_url"])
                current_info.append("virtual")
                current_info.append("virtual")
                current_info.append("virtual")
                current_info.append("virtual")
            else:
                if info_dict["venue"].get("name") is not None:
                    current_info.append(info_dict["venue"]["name"])
                else:
                    current_info.append("noname")
                current_info.append(info_dict["event_url"])
                if info_dict["venue"]["lat"] is not None:
                    current_info.append(info_dict["venue"]["lat"])
                else:
                    current_info.append("nolat")

                if info_dict["venue"]["lon"] is not None:
                    current_info.append(info_dict["venue"]["lon"])
                else:
                    current_info.append("nolon")

                if info_dict["venue"].get("country") is not None:
                    current_info.append(info_dict["venue"]["country"])
                else:
                    current_info.append("nocountry")

                if info_dict["venue"].get("city") is not None:
                    current_info.append(info_dict["venue"]["city"])
                else:
                    current_info.append("nocity")
            current_info.append(info_dict["group"]["id"])
            current_info.append(info_dict["created"])
            current_info.append(info_dict["visibility"])
            current_info.append(info_dict["time"])

            for index in range(len(current_info)):
                current_info[index] = FormatInfoString(current_info[index])
            csvfile.write("\"" + "\", \"".join(current_info) + "\"" + "\n")


    print("Extracting hosts")
    file_index = 1
    csv_heading = ["eventid", "memberid"]
    with open("../data/eventhosts.csv", "w") as csvfile:
        csvfile.write(", ".join(csv_heading) + "\n")

        for filename in file_list:
            print(f"Dealing with {file_index}/{file_number} ...", end="\r")
            file_index += 1
            host_list = []
            info_dict = ReadInfoFromEvent(filename)
            if info_dict.get("event_hosts") is not None:
                hosts = info_dict["event_hosts"]
                if isinstance(hosts["event_hosts_item"], list):
                    for item in hosts["event_hosts_item"]:
                        host_list.append(item["member_id"])
                else:
                    host_list=[hosts["event_hosts_item"]["member_id"]]
            for host_id in host_list:
                event_id = info_dict["id"]
                csvfile.write(f"{event_id}, {host_id}\n")



PrintcsvEvents()
```

```python
def PrintcsvRSVPs():
    group_info_file_list = "../data/rsvp.txt"

    print("Printing out Events")
    file_list = []
    with open(group_info_file_list, "r") as f:
        for line in f.readlines():
            line = line.strip()
            file_list.append(line)
    file_index = 1
    file_number = len(file_list)

    csv_heading = [
        "eventid",
        "memberid",
        "mtime"
    ]

    with open("../data/rsvp.csv", "w") as csvfile:
        csvfile.write(", ".join(csv_heading) + "\n")

        for filename in file_list:
            print(f"Dealing with {file_index}/{file_number} ...", end="\r")
            file_index += 1
            info_dict = ReadInfoFromRSVP(filename)

            if info_dict is None:
                continue
            if info_dict.get("event"):
                eventid = info_dict["event"]["id"]
            else:
                eventid = -1

            memberinfo = info_dict["memberinfo"]
            for item in memberinfo:
                memberid = item["member_id"]
                mtime = item["mtime"]
                csvfile.write(f"{eventid}, {memberid}, {mtime}" + "\n")

PrintcsvRSVPs()
```
