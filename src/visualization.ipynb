```python
user_data = pd.read_csv("../data/membertopics.csv")

info_count = user_data.shape[0]

# Scan1 : Find all topisz
topic_count = {}
user_count = {}
plot_data = {} # Final data
for index in range(info_count):
    user_id  = user_data.iloc[index][0]
    topic_id = user_data.iloc[index][1]
    if user_count.get(user_id):
        user_count[user_id].append(topic_id)
    else:
        user_count[user_id] = [topic_id]

    if topic_count.get(topic_id):
        topic_count[topic_id] += 1
    else:
        topic_count[topic_id] = 1

topic_count = dict(sorted(topic_count.items(), key=lambda item: -item[1]))
```
